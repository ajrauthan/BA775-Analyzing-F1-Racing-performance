### Section 1: Driver Analysis

#### Q1.1) What are the best drivers in the sport each decade?

```python
%%bigquery
-- creating a table to make buckets for each decade
SELECT
   year,
   CASE
      WHEN year BETWEEN 1950 AND 1959 THEN '1950s'
      WHEN year BETWEEN 1960 AND 1969 THEN '1960s'
      WHEN year BETWEEN 1970 AND 1979 THEN '1970s'
      WHEN year BETWEEN 1980 AND 1989 THEN '1980s'
      WHEN year BETWEEN 1990 AND 1999 THEN '1990s'
      WHEN year BETWEEN 2000 AND 2009 THEN '2000s'
      WHEN year BETWEEN 2010 AND 2019 THEN '2010s'
      WHEN year BETWEEN 2020 AND 2023 THEN '2020s'
   END AS era
FROM `ba775-a02-fall23.F1.seasons`;
```

```python
%%bigquery
WITH tab1 AS (
    SELECT
        era,
        CONCAT(forename, ' ', surname) AS name,
        SUM(points) AS points
    FROM
        `ba775-a02-fall23.F1.races` races
    JOIN
        `ba775-a02-fall23.F1.year_and_era` year
    ON
        races.year = year.year
    JOIN
        `ba775-a02-fall23.F1.driver_standings` driver_s
    ON
        races.raceId = driver_s.raceId
    JOIN
        `ba775-a02-fall23.F1.drivers` drivers
    ON
        drivers.driverId = driver_s.driverId
    GROUP BY
        era, name
)
-- creating a new table to find total points of each driver in each era
SELECT
    tab1.era,
    name
FROM
    (
        SELECT
            era,
            MAX(points) AS points
        FROM
            tab1
        GROUP BY
            era
    ) tab2
-- selecting the maximum points from each era
JOIN
    tab1
ON
    tab1.points = tab2.points
ORDER BY
    era DESC
-- joining tab 1 and tab 2 to show drivers with minimum points based on each era
```

#### Answer 1.1:
##### From the above table it is evident that some of these names are really familiar, with some of the legends of the sports being highlighted each decade. Looking into each of the eras:
##### 1950s: Juan Fangio, went on to win a total of 5 world championship titles in the years, 1951, 1954, 1955, 1956, and 1957.
##### 1960s: Graham Hill, went on to win a total of 2 world championship titles in the years, 1962, and 1968.
##### 1970s: Niki Lauda, went on to win a total of 3 world championship titles in the years, 1975, 1977, and the last championship title in the next decade in the year 1984.
##### 1980s: Alain Prost, went on to win a total of 4 world championship titles in the years, 1985, 1986, 1989, and the last championship title in the next decade in the year 1993.
##### 1990s & 2000s: Michael Schumacher, went on to win a total of 7 world championship titles across two decades, in the years, 1994, 1995, 2000, 2001, 2002, 2003, and 2004.
##### 2010s: Lewis Hamilton, went on to win a total of 7 world championship titles with his first being in the year 2008, and the later championship titles in 2014, 2015, 2017, 2018, 2019, 2020.
##### 2020s: Max Verstappen, went on to win a total of 3 world championship titles in the years 2021, 2022, 2023.

#### Q1.2) Who are the top 3 performing drivers from the previous season (2022)?

```python
%%bigquery
WITH top3_drivers AS (
    SELECT
        driverId,
        MAX(points) AS points
    FROM
        `ba775-a02-fall23.F1.driver_standings` ds
    JOIN
        `ba775-a02-fall23.F1.races` races ON ds.raceId = races.raceId
    WHERE
        year = 2022
    GROUP BY
        driverId
    ORDER BY
        points DESC
    LIMIT 3
)
-- creating a table with top 3 drivers based on the points they have accumulated in 2022
SELECT
    CONCAT(forename, ' ', surname) AS driver_name,
    points
FROM
    top3_drivers
LEFT JOIN
    `ba775-a02-fall23.F1.drivers` drivers ON top3_drivers.driverId = drivers.driverId
-- displaying the full names of top 3 drivers along with their respective points
```

#### Answer 1.2:
##### From the previous season of 2022, Max Verstappen won the World Championship, followed by Charles Leclerc coming in at 2nd and Sergio Perez getting the 3rd place. And it is interesting to note that the difference in points between the 2nd and 3rd place only had a difference of 3 points.

#### Q1.3) Which drivers have the most improvement a season in terms of their lap times, between the years of 2021 and 2022?

```python
%%bigquery
WITH PreviousSeasonLapTimes AS (
    SELECT
        lt.driverId,
        lt.lap,
        lt.milliseconds AS previousLapTime,
        r.year AS previousSeason
    FROM
        `ba775-a02-fall23.F1.lap_times` AS lt
    JOIN
        `ba775-a02-fall23.F1.races` AS r ON lt.raceId = r.raceId
    WHERE
        year = 2021
), -- creating a table to compute lap times of drivers in 2021
CurrentSeasonLapTimes AS (
    SELECT
        lt.driverId,
        lt.lap,
        lt.milliseconds AS currentLapTime,
        r.year AS currentSeason
    FROM
        `ba775-a02-fall23.F1.lap_times` AS lt
    JOIN
        `ba775-a02-fall23.F1.races` AS r ON lt.raceId = r.raceId
    WHERE
        year = 2022
) -- creating a table to compute lap times of drivers in 2022
SELECT
    CONCAT(d.forename, d.surname) AS name,
    AVG(c.currentLapTime) AS currentLapTime,
    AVG(p.previousLapTime) AS previousLapTime,
    (AVG(c.currentLapTime) - AVG(p.previousLapTime)) / 1000 AS improvement_seconds -- calculating the difference between lap times of the two seasons to track improvement
FROM
    CurrentSeasonLapTimes c
JOIN
    PreviousSeasonLapTimes p ON c.driverId = p.driverId AND c.lap = p.lap + 1
JOIN
    `ba775-a02-fall23.F1.drivers` AS d ON c.driverId = d.driverId
GROUP BY
    name
ORDER BY
    improvement_seconds DESC
LIMIT
    10;
```

#### Answer 1.3:
##### Building on our previous query, it is an interesting inference that Max Verstappen went on to really over do himself and perform his sbest increasing his overall laptime across races by a total of 4.6 seconds which goes along a long way to show the potential he has in this sport to do wonders and break more records.
##### It is also interesting to note that Sergio Perez saw an overall improvement of 3.8 seconds which points directly towards him coming 3rd in the 2022 season.

#### Q1.4) What are the top performing drivers in terms of in each of the race countries?

```python
%%bigquery
WITH BestPerformingDriverPerCountry AS (
    SELECT
        -- Use RANK() to assign a rank to each driver within their country based on total points
        RANK() OVER (PARTITION BY c.country ORDER BY SUM(re.points) DESC) AS rank,
        d.nationality,
        c.country AS race_country,
        d.driverID,
        d.forename,
        d.surname,
        -- Calculate the total points for each driver in each country
        SUM(re.points) AS total_points
    FROM
        -- Joining the tables races, circuits, and drivers
        `ba775-a02-fall23.F1.results` re
    JOIN
        `ba775-a02-fall23.F1.races` r ON re.raceID = r.raceID
    JOIN
        `ba775-a02-fall23.F1.Circuits` c ON r.circuitID = c.circuitID
    JOIN
        `ba775-a02-fall23.F1.drivers` d ON re.driverID = d.driverID
    -- Grouping the results by nationality, country, driverID, forename, and surname
    GROUP BY
        d.nationality, c.country, d.driverID, d.forename, d.surname
)
-- Fetching information from CTE for the best-performing driver in each country
SELECT
    nationality,
    race_country,
    driverID,
    forename,
    surname,
    total_points
FROM
    BestPerformingDriverPerCountry
-- Filtering to include only the top-ranked drivers in each country
WHERE
    rank = 1
-- Ordering by total points in descending order
ORDER BY total_points DESC;
```

#### Answer 1.4:
Going over the above table it is interesting to note that, apart from Lewis Hamilton dominiating in most of the countries including his home track of UK, while taking a closer look, it is interesting to note that players have the most number of points while racing on their home track.

For instance, Max Verstappen is Dutch and races for Red Bull who's home ground is in Austria, for which Max has accumulated the most number of points in Netherlands totaling up to 76 points and a total of 184 points in Austria.

Same goes for Michael Schumacher and Juan Fangio, who have accumulated the most number of points in their home track of Germany and Argentina, with the total number of points being 155 and 34 respectively.

This just goes to show how much home crowd goes on to support a drivers race attitude.

### Section 2: Constructor (Team) Analysis

#### 2.1) What are the top performing constructors since 2000?

```python
%%bigquery
SELECT
 races.raceId, year, races.name AS grand_prix, constructor_team,points, time
FROM
 `ba775-a02-fall23.F1.races` AS races
INNER JOIN
(
 SELECT
   constructors.constructorId, name AS constructor_team, nationality,
   constructor_standings.raceId, constructor_standings.points
 FROM
   `ba775-a02-fall23.F1.constructors` AS constructors
 INNER JOIN
   `ba775-a02-fall23.F1.constructor_standings` AS constructor_standings
 ON constructors.constructorId = constructor_standings.constructorId
) AS constructor_points
ON
 races.raceId = constructor_points.raceId
WHERE
 year>2000
ORDER BY
 points DESC
LIMIT 5;
```

##### Answer 2.1:
As we can see there are 2 major teams performing consistently since 2000, one being Mercedes and the other being Red Bull. Another interesting factors to notice is that Red Bull is present for the 2023 season even though it is just wrapping up as we type this.

#### Q2.2) What are the top performing constructors since 2000, that are unique?

```python
%%bigquery
SELECT constructor_team, SUM(points) as tot_points
FROM `ba775-a02-fall23.F1.races` AS races
INNER JOIN
(
 SELECT
   constructors.constructorId, name AS constructor_team, nationality,
   constructor_standings.raceId, constructor_standings.points
 FROM
   `ba775-a02-fall23.F1.constructors` AS constructors
 INNER JOIN
   `ba775-a02-fall23.F1.constructor_standings` AS constructor_standings
 ON constructors.constructorId = constructor_standings.constructorId
) AS constructor_points -- creating a table to compute total points for each constructor
ON races.raceId = constructor_points.raceId
WHERE year>2000
GROUP BY constructor_team
ORDER BY tot_points DESC
LIMIT 5;
```

##### Answer 2.2:
From the above query we can see a more even spread of constructors since 2000, with Mercedes having a total of 75542.5, followed by Ferrari having a total of 73128.5, followed by Red Bull having a total of 72680.5, followed by McLaren having a total of 38820.0, and finally Williams with a total of 15679.

While the top 3 performing teams have accumulated a total of 70000+ points, it is interesting to note that there is a significant drop in the points total preceeding that where it drops down to a range of 38000, followed by Williams at the range of 15000.

This just goes to show the kind of dominance certains teams have had over the past few years due to them receiving additional funding and in order to combat this the Federation Internationale de l'Automobile (FIA), in the past 2 years have set up rules to limit team spending in order to have a more competetive racing field across teams.

#### Q2.3) What are the top performing constructors in terms of the average pitstop time?

```python
%%bigquery
SELECT
  c.name as constructorName,  -- Select the constructor name
  AVG(CAST(ps.duration AS FLOAT64)) as avgPitStopTime  -- Calculate the average pit stop time, converting the duration to a numeric type
FROM
  `ba775-a02-fall23.F1.pit_stops` ps  -- From the pit_stops table
JOIN
  `ba775-a02-fall23.F1.results` r ON ps.raceId = r.raceId AND ps.driverId = r.driverId  -- Join with the results table using raceId and driverId to link pit stops to race results
JOIN
  `ba775-a02-fall23.F1.constructors` c ON r.constructorId = c.constructorId  -- Further join with the constructors table using constructorId to relate results to constructors
WHERE
  SAFE_CAST(ps.duration AS FLOAT64) IS NOT NULL  -- Ensure that duration is a valid numeric value before averaging
GROUP BY
  c.name  -- Group the results by constructor name
ORDER BY
  avgPitStopTime;  -- Order the results by the average pit stop time in ascending order
```

#### Answer 2.3)
##### Considering the results from our above query, it points directly towards the overall performance of the constructors, Red Bull, Mercedes & Ferrari have managed to be in the top 3 position for the average pit stop time.
##### Pit stops are one of the major factors that allows constructors to get an edge over each other apart from their cars performance.
##### It is also interesting to note that defunct teams as of today still hold records for the best average pit stop time, for instance Force India coming in at 4th andd Virgin coming in at 6th.

#### Q2.4) Who are the constuctors that are oldest in the sport?

```python
%%bigquery
SELECT c.name, (MAX(year) - MIN(year)) as duration -- calculating duration based on difference of minimum and maximum year for each constructor
FROM `ba775-a02-fall23.F1.constructor_standings` cs
JOIN `ba775-a02-fall23.F1.constructors` c ON cs.constructorId = c.constructorId
JOIN `ba775-a02-fall23.F1.races` races ON cs.raceId = races.raceId
GROUP BY name
ORDER BY duration DESC
LIMIT 10
```

#### Answer 2.4:
##### Ferrari is definitely a team that is infamous to the sport and it is staying true to it's legacy by being one of the oldest teams to date to be in the sport by being there for 65 years, followed by Aston Martin with a year younger at 64 and Alfa Romeo being in the sport for 60 years.
##### While not all of these teams are part of the sport still, teams like McLaren, Williams, still do compete in the sport with a legacy of 55 and 48 years.

### Section 3: Circuit  Analysis

#### Q3.1) Which circuits have hosted the most Formula 1 races?

```python
%%bigquery
SELECT
   c.name AS circuit_name,
   c.country AS circuit_country,
   COUNT(r.circuitId) AS number_of_races
FROM
   `ba775-a02-fall23.F1.Circuits` c
JOIN
   `ba775-a02-fall23.F1.races` r ON c.circuitId = r.circuitId
GROUP BY
   c.name, c.country  -- Grouping by both circuit name and country
ORDER BY
   number_of_races DESC
    LIMIT 10;
```

#### Q3.2) What is the oldest circuit in the history of Formula 1?

```python
%%bigquery
SELECT
  c.circuitId,
  c.name AS circuit_name,
  MIN(r.date) AS first_race_date,   -- using MIN function to select the earliest race date
  COUNT(r.raceId) AS total_races    -- using COUNT function to sum up the total number of races for each circuit
FROM
  `ba775-a02-fall23.F1.Circuits` c
JOIN
  `ba775-a02-fall23.F1.races` r ON c.circuitId = r.circuitId    -- joining circuits and races table using circuitID column
GROUP BY
  c.circuitId, c.name   -- grouping by circuit id and circuit name
ORDER BY
  first_race_date  -- ordering by earliest(ascending) race date
    LIMIT 10;
```

#### Answer 3.2:
##### F1 having a lot of history to the sport, it is very interesting note that one of the oldest circuits, Silverstone is still being used to date in the Grand Prix, followed by Monaco, Circuit de-Spa, Monza, being used in the Grand Prix to date.
##### Nürburgring in Germany while being one of the most popular racing tracks, the total number of races while being high with 41 races, it has seen a decline due a decline in German interest towards the sport.

#### Q3.3) Which circuits are most prone to accidents?

```python
%%bigquery
SELECT  name, COUNT(status) as n_mishaps -- counting number of mishaps to quantify the number of accidents and collisions taking place for each circuit
FROM
(
SELECT *
FROM `ba775-a02-fall23.F1.status` status
JOIN `ba775-a02-fall23.F1.results` results
ON status.statusId = results.statusId
JOIN `ba775-a02-fall23.F1.races` as races
ON results.raceId = races.raceId
)
WHERE status = 'Accident' OR status = 'Collision' -- filtering for mishaps based on the status column
GROUP BY name
ORDER BY n_mishaps DESC
LIMIT 10
```

#### Answer 3.3:
##### F1 as a sport has come a very long way with various safety precautions being put in place to avoid any dire situations, but certain track with how they are laid out, collisions or accidents are inevitable.
##### The track with the most mishaps being Monaco, is of no surprise as it is regarded as one of the most challenging tracks for the most talented of drivers, it is a street circuit with very narrow roads and sharp turns, and hence why it is ranked at the highest with 191 mishaps.
